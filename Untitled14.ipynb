The trimmed mean removes a certain percentage of the smallest and largest values before calculating the mean. For a 10% trimmed mean:

\text{Trimmed Mean} = \text{Mean of numbers excluding the lowest and highest 10% of values}

````python
import random
import statistics
from scipy.stats import trim_mean
from math import prod

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Mean
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# Median
def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 0:
        return (sorted_numbers[n // 2 - 1] + sorted_numbers[n // 2]) / 2
    else:
        return sorted_numbers[n // 2]

# Mode
def calculate_mode(numbers):
    try:
        return statistics.mode(numbers)
    except statistics.StatisticsError:
        return "No unique mode found"

# Weighted Mean
def calculate_weighted_mean(values, weights):
    return sum(v * w for v, w in zip(values, weights)) / sum(weights)

# Geometric Mean
def calculate_geometric_mean(numbers):
    return prod(numbers) ** (1 / len(numbers))

# Harmonic Mean
def calculate_harmonic_mean(numbers):
    return len(numbers) / sum(1 / x for x in numbers)

# Midrange
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# Trimmed Mean (excluding a percentage of outliers)
def calculate_trimmed_mean(numbers, percentage):
    trim_fraction = percentage / 100.0
    return trim_mean(numbers, trim_fraction)

# Generate random weights for weighted mean
weights = [random.uniform(0.5, 1.5) for _ in range(100)]

# Results
mean = calculate_mean(int_list)
median = calculate_median(int_list)
mode = calculate_mode(int_list)
weighted_mean = calculate_weighted_mean(int_list, weights)
geometric_mean = calculate_geometric_mean(int_list)
harmonic_mean = calculate_harmonic_mean(int_list)
midrange = calculate_midrange(int_list)
trimmed_mean_10 = calculate_trimmed_mean(int_list, 10)  # 10% trimmed mean

# Outputs
{
    "Mean": mean,
    "Median": median,
    "Mode": mode,
    "Weighted Mean": weighted_mean,
    "Geometric Mean": geometric_mean,
    "Harmonic Mean": harmonic_mean,
    "Midrange": midrange,
    "Trimmed Mean (10%)": trimmed_mean_10
}
````

3)  Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance.

````python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """
        Initialize the discrete random variable.

        :param values: List of possible values of the random variable.
        :param probabilities: List of probabilities associated with each value.
        """
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not sum(probabilities) == 1:
            raise ValueError("The probabilities must sum to 1.")

        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value (mean) of the random variable.

        :return: Expected value (float).
        """
        return sum(v * p for v, p in zip(self.values, self.probabilities))

    def variance(self):
        """
        Calculate the variance of the random variable.

        :return: Variance (float).
        """
        mean = self.expected_value()
        return sum(p * (v - mean) ** 2 for v, p in zip(self.values, self.probabilities))


# Example Usage
values = [1, 2, 3, 4, 5]  # Possible outcomes
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]  # Corresponding probabilities

# Create an instance of the random variable
random_variable = DiscreteRandomVariable(values, probabilities)

# Calculate expected value and variance
expected_value = random_variable.expected_value()
variance = random_variable.variance()

print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")
````

4) Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and
variance of the outcomes.


 Simulating the Rolling of a Fair Six-Sided Die
The expected value and variance for a fair six-sided die can be calculated theoretically, but here we simulate the rolls and compute these metrics programmatically.

````python
import random

def roll_die(num_rolls):
    """
    Simulate rolling a fair six-sided die.

    :param num_rolls: Number of rolls to simulate.
    :return: List of outcomes.
    """
    return [random.randint(1, 6) for _ in range(num_rolls)]

def calculate_expected_value_and_variance(outcomes):
    """
    Calculate the expected value and variance of die roll outcomes.

    :param outcomes: List of outcomes.
    :return: Expected value and variance.
    """
    n = len(outcomes)
    mean = sum(outcomes) / n
    variance = sum((x - mean) ** 2 for x in outcomes) / n
    return mean, variance

# Simulate rolling a die 10,000 times
num_rolls = 10000
outcomes = roll_die(num_rolls)
expected_value, variance = calculate_expected_value_and_variance(outcomes)

print(f"Simulated Expected Value: {expected_value}")
print(f"Simulated Variance: {variance}")
````

5) Create a Python function to generate random samples from a given probability distribution (e.g.,
binomial, Poisson) and calculate their mean and variance.

````python
import numpy as np

def generate_random_samples(distribution, params, num_samples):
    """
    Generate random samples from a given probability distribution.

    :param distribution: Name of the distribution ('binomial' or 'poisson').
    :param params: Parameters for the distribution (e.g., n and p for binomial).
    :param num_samples: Number of samples to generate.
    :return: List of generated samples.
    """
    if distribution == 'binomial':
        n, p = params
        return np.random.binomial(n, p, num_samples)
    elif distribution == 'poisson':
        lam = params[0]
        return np.random.poisson(lam, num_samples)
    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")

def calculate_mean_and_variance(samples):
    """
    Calculate the mean and variance of the generated samples.

    :param samples: List of samples.
    :return: Mean and variance.
    """
    mean = np.mean(samples)
    variance = np.var(samples)
    return mean, variance

# Example usage for binomial distribution
num_samples = 10000
binomial_samples = generate_random_samples('binomial', (10, 0.5), num_samples)
binomial_mean, binomial_variance = calculate_mean_and_variance(binomial_samples)

print(f"Binomial Distribution - Mean: {binomial_mean}, Variance: {binomial_variance}")

# Example usage for Poisson distribution
poisson_samples = generate_random_samples('poisson', (3,), num_samples)
poisson_mean, poisson_variance = calculate_mean_and_variance(poisson_samples)

print(f"Poisson Distribution - Mean: {poisson_mean}, Variance: {poisson_variance}")
````

6) Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute
the mean, variance, and standard deviation of the samples.

````python
import numpy as np

def generate_gaussian_samples(mean, std_dev, num_samples):
    """
    Generate random numbers from a Gaussian (normal) distribution.

    :param mean: Mean of the Gaussian distribution.
    :param std_dev: Standard deviation of the Gaussian distribution.
    :param num_samples: Number of samples to generate.
    :return: List of generated samples.
    """
    return np.random.normal(mean, std_dev, num_samples)

def compute_statistics(samples):
    """
    Compute the mean, variance, and standard deviation of the samples.

    :param samples: List of samples.
    :return: Mean, variance, and standard deviation.
    """
    mean = np.mean(samples)
    variance = np.var(samples)
    std_dev = np.std(samples)
    return mean, variance, std_dev

# Example usage
num_samples = 10000
mean = 50  # Mean of the Gaussian distribution
std_dev = 10  # Standard deviation of the Gaussian distribution

# Generate Gaussian samples
samples = generate_gaussian_samples(mean, std_dev, num_samples)

# Compute statistics
sample_mean, sample_variance, sample_std_dev = compute_statistics(samples)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
print(f"Sample Standard Deviation: {sample_std_dev}")
````

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")
````

Calculate Skewness

````python
def calculate_skewness(column):
    """
    Calculate the skewness of a column.

    :param column: Pandas Series or a list of numbers.
    :return: Skewness value (float).
    """
    mean = column.mean()
    median = column.median()
    std_dev = column.std()
    skewness = (3 * (mean - median)) / std_dev
    return skewness

# Calculate skewness for total_bill and tip
total_bill_skewness = calculate_skewness(tips["total_bill"])
tip_skewness = calculate_skewness(tips["tip"])

print(f"Skewness of total_bill: {total_bill_skewness}")
print(f"Skewness of tip: {tip_skewness}")
````

 Determine Skewness Type

````python
def classify_skewness(skewness):
    """
    Classify the skewness as positive, negative, or symmetric.

    :param skewness: Skewness value (float).
    :return: Skewness type (string).
    """
    if skewness > 0:
        return "Positive Skewness"
    elif skewness < 0:
        return "Negative Skewness"
    else:
        return "Approximately Symmetric"

# Determine skewness type
total_bill_skewness_type = classify_skewness(total_bill_skewness)
tip_skewness_type = classify_skewness(tip_skewness)

print(f"Total Bill Skewness Type: {total_bill_skewness_type}")
print(f"Tip Skewness Type: {tip_skewness_type}")
````

Calculate Covariance

````python
def calculate_covariance(column1, column2):
    """
    Calculate the covariance between two columns.

    :param column1: Pandas Series or list.
    :param column2: Pandas Series or list.
    :return: Covariance value (float).
    """
    covariance_matrix = np.cov(column1, column2)
    return covariance_matrix[0, 1]

# Calculate covariance between total_bill and tip
covariance = calculate_covariance(tips["total_bill"], tips["tip"])
print(f"Covariance between total_bill and tip: {covariance}")
````

Calculate Pearson Correlation Coefficient

````python
def calculate_correlation(column1, column2):
    """
    Calculate the Pearson correlation coefficient between two columns.

    :param column1: Pandas Series or list.
    :param column2: Pandas Series or list.
    :return: Pearson correlation coefficient (float).
    """
    return np.corrcoef(column1, column2)[0, 1]

# Calculate Pearson correlation between total_bill and tip
correlation = calculate_correlation(tips["total_bill"], tips["tip"])
print(f"Pearson Correlation Coefficient between total_bill and tip: {correlation}")
````

````python
from scipy.stats import norm

def calculate_normal_pdf(x, mean, std_dev):
    """
    Calculate the Probability Density Function (PDF) for a given normal distribution.

    :param x: Value for which PDF is calculated.
    :param mean: Mean of the normal distribution.
    :param std_dev: Standard deviation of the normal distribution.
    :return: PDF value (float).
    """
    return norm.pdf(x, loc=mean, scale=std_dev)

# Example usage
mean = 0
std_dev = 1
x_value = 1
pdf_value = calculate_normal_pdf(x_value, mean, std_dev)
print(f"PDF value at x = {x_value} for N({mean}, {std_dev**2}): {pdf_value}")
````

````python
from scipy.stats import poisson

def calculate_poisson_pmf(k, lam):
    """
    Calculate the Probability Mass Function (PMF) for a Poisson distribution.

    :param k: Value for which PMF is calculated (number of occurrences).
    :param lam: Average rate of occurrence (lambda).
    :return: PMF value (float).
    """
    return poisson.pmf(k, mu=lam)

# Example usage
lam = 3  # Î» (average rate of occurrence)
k_value = 2
pmf_value = calculate_poisson_pmf(k_value, lam)
print(f"PMF value for k = {k_value} with Î» = {lam}: {pmf_value}")
````

  ```python

  pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]

  post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

````python
import numpy as np
from scipy import stats

def paired_t_test(pre_scores, post_scores):
    # Ensure both arrays have the same length
    if len(pre_scores) != len(post_scores):
        raise ValueError("Both input arrays must have the same length.")

    # Calculate the differences between the paired scores
    differences = np.array(post_scores) - np.array(pre_scores)

    # Calculate the mean and standard deviation of the differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)
    n = len(differences)

    # Calculate the t-statistic
    t_stat = mean_diff / (std_diff / np.sqrt(n))

    # Calculate degrees of freedom
    df = n - 1

    # Calculate the p-value for a one-tailed test (since we expect post > pre)
    p_value = stats.t.cdf(t_stat, df)

    # Output results
    print(f"Mean of differences: {mean_diff:.2f}")
    print(f"Standard Deviation of differences: {std_diff:.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value:.4f}")

    # Determine if the result is significant at the 0.05 level
    alpha = 0.05
    if p_value < alpha:
        print("Reject null hypothesis: The intervention had a significant positive impact on test scores.")
    else:
        print("Fail to reject null hypothesis: No significant impact from the intervention.")

# Test data for pre- and post-intervention scores
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform the paired t-test
paired_t_test(pre_intervention_scores, post_intervention_scores)
````

````python
import numpy as np
from scipy import stats

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

def two_sample_t_test(male_data, female_data):
    # Calculate the sample means and standard deviations
    mean_male = np.mean(male_data)
    mean_female = np.mean(female_data)

    std_male = np.std(male_data, ddof=1)
    std_female = np.std(female_data, ddof=1)

    n_male = len(male_data)
    n_female = len(female_data)

    # Calculate the t-statistic
    t_stat = (mean_male - mean_female) / np.sqrt((std_male**2 / n_male) + (std_female**2 / n_female))

    # Calculate the degrees of freedom using Welchâ€™s formula
    numerator = (std_male**2 / n_male + std_female**2 / n_female)**2
    denominator = ((std_male**2 / n_male)**2 / (n_male - 1)) + ((std_female**2 / n_female)**2 / (n_female - 1))
    df = numerator / denominator

    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

    # Output the results
    print(f"Mean male salary: {mean_male:.2f}")
    print(f"Mean female salary: {mean_female:.2f}")
    print(f"Standard Deviation (Male): {std_male:.2f}")
    print(f"Standard Deviation (Female): {std_female:.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"Degrees of Freedom: {df:.3f}")
    print(f"P-value: {p_value:.4f}")

    # Check if the result is statistically significant at alpha = 0.05
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference in average salaries.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in average salaries.")

# Perform the two-sample t-test
two_sample_t_test(male_salaries, female_salaries)
````

21) A political analyst wants to determine if there is a significant association between age groups and voter
preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.


Use the below code to generate data:

```python

np.random.seed(0)

age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

````python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate synthetic data
np.random.seed(0)

# Define age groups and voter preferences
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform the Chi-Square Test of Independence
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Output the results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_val:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")

# Determine if the result is statistically significant
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis: No significant association between age groups and voter preferences.")
````
